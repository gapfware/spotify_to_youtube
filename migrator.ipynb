```python
from dotenv import load_dotenv
import spotipy
import itertools 
from spotipy.oauth2 import SpotifyClientCredentials
from ytmusicapi import YTMusic
```

```python
load_dotenv()
```

```python
auth_manager = SpotifyClientCredentials()
spotify_client = spotipy.Spotify(auth_manager=auth_manager)
ytmusic = YTMusic("oauth.json")
```

```python
def playlist_items(spotify_playlist_id):
    for offset in itertools.count(step=100):
        response = spotify_client.playlist_items(spotify_playlist_id, offset=offset, limit=100)
        for item in response["items"]:
            yield item
        if response["next"] is None:
            return
```

```python

sp_playlist = list(playlist_items("1NDxNw7bDudaEaM1qqK2cc"))
playlist_title = spotify_client.playlist("1NDxNw7bDudaEaM1qqK2cc")['name']
print(type(playlist_title))
```

```python
yt_playlist_id = ytmusic.create_playlist("x191 migration", description="Creado desde python")
```

```python
videos_id = []
for item in sp_playlist:
    title = item["track"]["name"]
    artists = " ".join(
        [
            artist["name"]
            for artist in item["track"]["artists"]
            if artist["type"] == "artist"
        ]
    )
    query = f"{title} {artists}"
    search_results = ytmusic.search(query=query)
    for result in search_results:
        if result["resultType"] == "song" and result not in videos_id:
            print(result["title"])
            videos_id.append(result["videoId"])
            break
    else:
        print(f"Could Not find a match to {query}")
```

```python
print(len(videos_id))
ids = set(videos_id)
```

```python
add = ytmusic.add_playlist_items(playlistId=yt_playlist_id, videoIds=ids)
if add['status'] == 'STATUS_FAILED':
    print('No se agregaron al playlist')
else:
    print('Migrado correctamente')
```

```python
from urllib.parse import urlparse
def obtener_ultima_parte_de_url(url):
    parsed_url = urlparse(url)
    path = parsed_url.path
    partes_del_path = path.split('/')
    ultima_parte = partes_del_path[-1] if partes_del_path else ''
    return ultima_parte

# Ejem
url = 'https://open.spotify.com/playlist/17TJaD0n9nJsgp9c5kbdLT'
ultima_parte = obtener_ultima_parte_de_url(url)
print("ltima parte de la URL:", ultima_parte)
```
