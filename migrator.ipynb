```python
from dotenv import load_dotenv
import spotipy
import itertools 
from spotipy.oauth2 import SpotifyClientCredentials
from ytmusicapi import YTMusic
```

```python
load_dotenv()
```

```python
auth_manager = SpotifyClientCredentials()
sp = spotipy.Spotify(auth_manager=auth_manager)
ytmusic = YTMusic("oauth.json")
```

```python
def playlist_items(spotify_playlist_id):
    for offset in itertools.count(step=100):
        response = sp.playlist_items(spotify_playlist_id, offset=offset, limit=100)
        for item in response["items"]:
            yield item
        if response["next"] is None:
            return
```

```python

sp_playlist = list(playlist_items("5oHRecaICRYutJnlWTcSV3"))
len(sp_playlist)
```

```python
yt_playlist_id = ytmusic.create_playlist("x191 migration", description="Creado desde python")
```

```python
videos_id = []
for item in sp_playlist:
    title = item["track"]["name"]
    artists = " ".join(
        [
            artist["name"]
            for artist in item["track"]["artists"]
            if artist["type"] == "artist"
        ]
    )
    query = f"{title} {artists}"
    search_results = ytmusic.search(query=query)
    for result in search_results:
        if result["resultType"] == "song" and result not in videos_id:
            print(result["title"])
            videos_id.append(result["videoId"])
            break
    else:
        print(f"Could Not find a match to {query}")
```

```python
ids = set(videos_id)
```

```python
ytmusic.add_playlist_items(playlistId=yt_playlist_id, videoIds=list(ids))
```
